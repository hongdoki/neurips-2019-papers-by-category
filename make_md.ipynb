```python
import requests 
URL = 'https://nips.cc/Conferences/2019/Schedule' 
response = requests.get(URL) 
response.status_code 
source = response.text
```

```python
from bs4 import BeautifulSoup
```

```python
bs = BeautifulSoup(source, 'html.parser')
```

```python
right_tags = bs.findAll('div', attrs={'class':'pull-right maincardHeader maincardType'})
```

```python
tus = list()
for rt in right_tags:
    if rt.find('a') is not None:
        link_tag = rt.find('a')
        title = link_tag.contents
        url = list(link_tag.attrs.values())[0]
        tus += [(title[0].lstrip(), url)]
```

```python
from collections import Counter
cnter = Counter(tus)
prev_cats = ['','']
md_list = list()
for tu in sorted(cnter.keys()):
    cnt = cnter[tu]
    cats = tu[0].split(' -- ')
    if len(cats) == 2:
        if cats[0] != prev_cats[0] : md_list += ['## ' + cats[0]]
        if cats[1] != prev_cats[1] : md_list += ['* ' + cats[1]]
        md_list += ['    * [%d papers](http://neurips.cc' % cnt + tu[1] + ')']
    elif len(cats) == 1:
        if cats[0] != prev_cats[0] : md_list += ['* ' + cats[0]]
        md_list += ['    * [%d papers](http://neurips.cc' % cnt + tu[1] + ')']
        cats += [cats[0]]
    prev_cats = cats
    
```

```python
with open('README.md', 'w') as f:
    f.write('\n'.join(md_list))
```
